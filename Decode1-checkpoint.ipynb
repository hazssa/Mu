```python
#filling miss value or object
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")

df.isnull().sum()
```

```python
#chcking for duplicate

duplicate_count = df.duplicated().sum()

print("Number of duplicate rows:", duplicate_count)
```

```python
#Convert Date Column Properly
df["Date"] = pd.to_datetime(df["Date"])

df.dtypes
```

```python
#Create New Date Columns

df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["MonthName"] = df["Date"].dt.month_name()
df["YearMonth"] = df["Date"].dt.to_period("M").astype(str)
df["DayName"] = df["Date"].dt.day_name()

df.head()
```

```python
#Validate Total Price
#Quantity * Unitprice = Totalprice
df["CalculatedTotal"] = df["Quantity"] * df["UnitPrice"]
df["PriceDifference"] = df["TotalPrice"] - df["CalculatedTotal"]

df[["Quantity", "UnitPrice", "TotalPrice", "CalculatedTotal", "PriceDifference"]].head()
```

```python
#Overall Business KPIs

total_revenue = df["TotalPrice"].sum()
total_orders = df["OrderID"].nunique()
total_customers = df["CustomerID"].nunique()
average_order_value = df["TotalPrice"].mean()
total_quantity_sold = df["Quantity"].sum()

print("Total Revenue:", round(total_revenue, 2))
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Average Order Value:", round(average_order_value, 2))
print("Total Quantity Sold:", total_quantity_sold)
```

```python
# Function to add values on bar charts

def add_bar_labels(ax, fontsize=8):
    max_height = 0

    for bar in ax.patches:
        height = bar.get_height()
        if height > max_height:
            max_height = height

    for bar in ax.patches:
        value = bar.get_height()

        if value > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                value + (max_height * 0.01),
                f"{value:,.0f}",
                ha="center",
                va="bottom",
                fontsize=fontsize
            )

    ax.set_ylim(0, max_height * 1.18)
```

```python
# Function to add values on line charts

def add_line_labels(ax, x_values, y_values, fontsize=7):
    max_value = max(y_values)

    for x, y in zip(x_values, y_values):
        ax.text(
            x,
            y + (max_value * 0.01),
            f"{y:,.0f}",
            ha="center",
            va="bottom",
            fontsize=fontsize
        )

    ax.set_ylim(0, max_value * 1.18)
```
